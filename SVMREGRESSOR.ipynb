```python
X=df.drop("price",axis=1)
y=df["price"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
siralama={
    "cut":['Fair', 'Good', 'Very Good', 'Premium', 'Ideal'],
     "color":['J', 'I', 'H', 'G', 'F', 'E', 'D'],
    "clarity":['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']
}
```

```python
encoder = OrdinalEncoder(categories=[siralama["cut"], siralama["color"], siralama["clarity"]])
```

```python
X_train[["cut", "color", "clarity"]] = encoder.fit_transform(X_train[["cut", "color", "clarity"]])
X_test[["cut", "color", "clarity"]]=encoder.transform(X_test[["cut", "color", "clarity"]])
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
scaler=RobustScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
y_pred=regression.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
plt.scatter(y_test,y_pred)
plt.show()
```
