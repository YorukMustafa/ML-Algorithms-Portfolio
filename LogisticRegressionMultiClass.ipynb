```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("7-cyber_attack_data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(columns="attack_type",axis=1)
y=df["attack_type"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
regression=LogisticRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
y_pred=regression.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
score=accuracy_score(y_test,y_pred)
classifiction_report=classification_report(y_test,y_pred)
matrix=confusion_matrix(y_test,y_pred)
print("score :",score,"\n classifiction_report :", classifiction_report, "\n confusion_matrix :" , matrix)
```

```python
from sklearn.multiclass import OneVsOneClassifier,OneVsRestClassifier
```

```python
onevsonecls=OneVsOneClassifier(LogisticRegression())
```

```python
onevsrestcls=OneVsRestClassifier(LogisticRegression())
```

```python
onevsonecls.fit(X_train,y_train)
```

```python
y_pred2=onevsonecls.predict(X_test)
```

```python
y_pred2
```

```python
score=accuracy_score(y_test,y_pred2)
classifiction_report=classification_report(y_test,y_pred2)
matrix=confusion_matrix(y_test,y_pred2)
print("score :",score,"\n classifiction_report :", classifiction_report, "\n confusion_matrix :" , matrix)
```
