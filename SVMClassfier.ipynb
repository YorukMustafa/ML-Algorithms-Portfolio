```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("9-email_classification_svm.csv")
```

```python
df.describe()
```

```python
df.info()
```

```python
sns.scatterplot(x="subject_formality_score",y="sender_relationship_score",data=df,hue="email_type")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(columns="email_type")
y=df["email_type"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
y_test
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel="linear")
```

```python
svc.fit(X_train,y_train)
```

```python
svc.coef_
```

```python
svc.intercept_
```

```python
y_pred=svc.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print("accuracy_score:",accuracy_score(y_test,y_pred))
```

```python
print("classification_report:",classification_report(y_test,y_pred))
print("confusion_matrix:",confusion_matrix(y_test,y_pred))
```

```python
sns.scatterplot(x="underground_wave_energy",y="vibration_axis_variation",hue="seismic_event_detected",data=df)
plt.show()
```

```python
X=df.drop("seismic_event_detected",axis=1)
y=df["seismic_event_detected"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 15)
```

```python




rbf=SVC(kernel='rbf')
rbf.fit(X_train,y_train)
y_pred11=rbf.predict(X_test)
print(classification_report(y_test,y_pred11))
print(confusion_matrix(y_test,y_pred11))



```

```python


linear=SVC(kernel='linear')
linear.fit(X_train,y_train)
y_pred9=linear.predict(X_test)
print(classification_report(y_test,y_pred9))
print(confusion_matrix(y_test,y_pred9))


```
