```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("6-bank_customers.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
X=df.drop(columns="subscribed",axis=1)
y=df["subscribed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic=LogisticRegression()
```

```python
logistic.fit(X_train,y_train)
```

```python
logistic.coef_
```

```python
logistic.intercept_
```

```python
y_pred1=logistic.predict(X_test)
```

```python
y_pred1
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy=accuracy_score(y_test,y_pred1)
confusion_mtrx=confusion_matrix(y_test,y_pred1)
classification=classification_report(y_test,y_pred1)
```

```python
print("accuracy:",accuracy)
print("confusion_mtrx:",confusion_mtrx)
print("classification:",classification)
```

```python
regression=LogisticRegression()
```

```python
penalyts=["l1","l2","elasticnet"]
c_values=[100,10,1,0.1]
solver=['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga','lbfgs']
dictionary=dict(C=c_values,penalty=penalyts,solver=solver)
```

```python
from sklearn.model_selection import GridSearchCV,StratifiedKFold
```

```python
cv=StratifiedKFold()
```

```python
grid=GridSearchCV(estimator=regression,param_grid=dictionary,cv=cv,scoring="accuracy",n_jobs=-1)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
grid.best_estimator_
```

```python
grid.classes_
```

```python
y_pred2=grid.predict(X_test)
```

```python
accuracy=accuracy_score(y_test,y_pred2)
confusion_mtrx=confusion_matrix(y_test,y_pred2)
classification=classification_report(y_test,y_pred2)
```

```python
print("accuracy:",accuracy)
print("confusion_mtrx:",confusion_mtrx)
print("classification:",classification)
```
